```python
sleeping=[1,2,3,4,6]
eating=[2,3,4,5,6]
working=[8,7,4,1,3]
playing=[8,6,2,3,5]

plt.plot([],[],label='sleeping',color='m',linewidth=10)
plt.plot([],[],label='eating',color='y',linewidth=10)
plt.plot([],[],label='working',color='r',linewidth=10)
plt.plot([],[],label='playing',color='b',linewidth=10)

plt.stackplot(sleeping,eating,working,playing,colors=['m','y','r','b'])
plt.legend()
```
