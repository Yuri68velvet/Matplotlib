```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,6,12)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y,'b--')

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('This is proportional glaph')

plt.show()
```

# Creating Multiplots on Same Canvas

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')

plt.subplot(1,2,2)
plt.plot(y,x,'b*-')

plt.show()
```

```python
fig=plt.figure()
axes=fig.add_axes([0.2,0.2,0.7,0.7])

axes.plot(x,y,'r')
axes.set_xlabel('X-AXIS')
axes.set_ylabel('Y-AXIS')
axes.set_title('TITLE')
```

```python
fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y,'b--')
axes.set_xlabel('Set X-label')
axes.set_ylabel('Set Y-label')

axes.set_title('Set title')
```
