# Bar plots

```python
x=[12,34,56,66,77]
y=[20,30,40,50,60]

x2=[1,2,3,4,5,6,7,8,9]
y2=[10,20,30,40,50,60,70,80,90]

plt.bar(x,y,label='new bar chart ')

plt.bar(x2,y2,label='old bar chart')

plt.legend()
plt.show()
```

```python
x=[12,34,56,66,77]
y=[20,30,40,50,60]

x2=[1,2,3,4,5,6,7,8,9]
y2=[10,20,30,40,50,60,70,80,90]

plt.bar(x,y,label='new bar chart ',color='r')
color='#D0D3D4'
plt.bar(x2,y2,label='old bar chart',color=color)

plt.legend()
plt.show()
```

# first plot is for bar tabs

```python
population_ages=[20,12,32,43,55,45,80,130,80,125,123,25,29,100]

ids=[x for x in range(len(population_ages))]

plt.bar(ids,population_ages,label='Graph')
plt.xlabel('ids for x')
plt.ylabel('population_ages')
plt.legend()
#plt.show()
```

```python
#if i use histogram
population_ages=[20,65,45,2,43,56,75,78,34,38,68,61,69,69,
80,120,101]
bins=[0,20,30,40,50,60,70,80,90,100,110,120,130]

plt.hist(population_ages,bins,histtype='bar',rwidth=0.8)
plt.xlabel('X-population_ages')

plt.ylabel('Y-bins')

plt.show()
```
