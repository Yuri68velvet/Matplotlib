# Basic Matplotlib Commands

# Matplotlib Object Method
Now that we've seen the  basics,let's break it all down with a more formal introduction of Matplotlib's Object Oriented API.This means we will
instantiate figure objects and then call methods or attributes from that object.

# Introduction to the Object Oriented Method

```python
#Create Figure(empty canvas)

fig=plt.figure()
#add set of axes to figure

axes=fig.add_axes([0.1,0.1,0.8,0.8])

#plot on that o axes

axes.plot(x,y,'b')
axes.set_xlabel('Set X-label')
axes.set_ylabel('Set Y-label')

axes.set_title('Set title')
```

#### Code is a little complicated, but the advantage is that we have full control of where the plot axs are placed,
and we ccan easily add more than one axis to the figure.

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])#main  axes
axes2=fig.add_axes([0.8,0.1,0.4,0.3])



#larger Figure Axes 1
axes1.plot(x,y,'b')
axes1.set_xlabel('X label for Large')
axes1.set_ylabel('Y la for Large')
axes1.set_title('Title for Larger')

axes2.plot(y,x,'r--')
axes2.set_xlabel('X label for Small')
axes2.set_ylabel('Y label for Small')
axes2.set_title('Title for Small')
```

# Subplots()
The plt.subplots() object will act as a more automatic axis manager.
Basic use cases:

```python
# Use similar to plt.figure() except use tuple unpaking to grab fig and axes

```

```python
fig,axes=plt.subplots()
#new use the axes object to add stuffs to plot

axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Title')
```

#### Then you can specify the number of rows and columns when creating the subplots() object:

```python
#Empty canvas of 1 by 2 subplots
fig,axes=plt.subplots(nrows=1,ncols=2)
```

```python
# Axes is an array of axes to plot on
axes
```

#### We can iterate through this array:

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

for ax in axes:
    ax.plot(x,y,'b')
    ax.set_xlabel('shdvo;')
    ax.set_ylabel('hau;vb')
    ax.set_title('Title')
    
fig
plt.tight_layout() 
```

which automatically adjusts the position of the axes on the figure canvas so that there is no overlapping content:

# Figure size,aspect ratio and DPI

#### Matplotlib allows the aspect ratio,DPI and figure size to be specified when the Figure object is created. 
You can ues the figsize and dpi keyword arguments.

* figsize is a tuple of the width and height of the figure in inches
* dpi is the dots-per-inch(pixel per inch).

For example:

```python
fig=plt.figure(figsize=(8,8),dpi=100)
```

#### The same arguments can also be passed to layout managers,such as the subplots function

```python
fig,axes=plt.subplots(figsize=(6,8))

axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Title')
```

# Legends,labels and title

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label="x**2")
ax.plot(x,x**3,label="x**3")
ax.legend(loc=1)#right sides upper=1
```
