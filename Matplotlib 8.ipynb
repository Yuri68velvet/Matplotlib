```python
import matplotlib.pyplot as plt
import csv
```

```python
x=[]
y=[]

#r read
with open('matplotlibnote.txt','r') as csvfile:
    plots=csv.reader(csvfile,delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
        
plt.plot(x,y,label='Load from the file',marker='*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('this ois Graph')
plt.legend()
```

```python
import numpy as np

x,y=np.loadtxt('matplotlibnote.txt',delimiter=',',unpack=True)

plt.plot(x,y,label="Easy way to Do",marker='*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('This is Graph')
plt.legend()
plt.show()
```
