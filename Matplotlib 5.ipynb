```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[5,3,5,7,2,3,4,7,1]
```

```python
plt.scatter(x,y)
plt.scatter(x,y,label='Scatter plot')
plt.legend()
```

```python
plt.scatter(x,y,color='k')
```

```python
plt.scatter(x,y,color='y',marker='v')
```

```python
plt.scatter(x,y,color='g',marker='D',s=120)
```
