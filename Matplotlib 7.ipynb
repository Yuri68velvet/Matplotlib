```python
import matplotlib.pyplot as plt
```

```python
sleeping=[7,8,9,11,7]
eating=[2,3,4,5,6]
working=[8,7,4,1,3]
playing=[8,6,2,3,5]
slices=[7,2,2,13]
activities=['sleeping','eating','working','playing']
colors=['m','y','b','r']
```

```python
plt.pie(slices,labels=activities,colors=colors)
```

```python
plt.pie(slices,labels=activities,colors=colors,startangle=90,shadow=True,explode=(0,0.1,0.3,0),autopct='%1.1f%%')
```
